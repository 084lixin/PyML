## 3.4. matplotlib的使用

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 5)
y = 2 * x + 1
```

#### 绘制散点图

```python
plt.scatter(x, y)
```

#### 调整尺寸

```python
plt.scatter(x, y, s=200)
```

#### 调整形状

```python
plt.scatter(x, y, s=200, marker=',')
```

#### 更改颜色

```python
plt.scatter(x, y, s=200, marker=',', c='red')
```

#### 更改坐标

```python
plt.xlim((0, 30))
plt.ylim((0, 50))

plt.scatter(x, y, s=200, marker=',', c='red')
```

#### 更改刻度

```python
plt.xlim((0, 30))
plt.ylim((0, 50))

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.scatter(x, y, s=200, marker=',', c='red')
```

#### 增加标签

```python
plt.xlim((0, 30))
plt.ylim((0, 50))

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.xlabel('x_label')
plt.ylabel('y_label')

plt.scatter(x, y, s=200, marker=',', c='red')
```

#### 绘制线

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
y0 = -x * x
plt.plot(x, y)
plt.plot(x, y0, c='red')
```

```python
x2 = np.linspace(-10, 10, 5)
y2 = 3 * x2 + 50

plt.scatter(x2, y2, c='red')
plt.plot(x, y)
```

#### 练习答案

```python
x = np.linspace(-5, 5, 50)
y = np.sin(x)
plt.xlim(-5,5)
plt.plot(x, y)
```

#### 柱状图

```python
x = [1, 2, 3]
y = [5, 3, 4]
plt.bar(x, y)
```

```python
plt.bar(x, y, width = 0.2, color = 'r')
```
