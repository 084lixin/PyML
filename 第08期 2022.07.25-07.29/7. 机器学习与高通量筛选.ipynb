```python
df_el = load_dataset('elastic_tensor_2015', data_home = '.')
df_el
```

```python
import matplotlib.pyplot as plt
plt.hist(df_el['K_VRH'])
plt.show()
```

```python
plt.hist(df_el['K_VRH'], color = 'g', edgecolor = 'k', alpha = 0.75)
plt.show()
```

```python
plt.hist(df_el['K_VRH'], color = 'g', edgecolor = 'k', alpha = 0.75,
        bins = [0, 50, 100, 150, 200, 250, 300, 350, 400])
plt.show()
```

```python
plt.hist(df_el['K_VRH'], density = True, color = 'g',
         edgecolor = 'k', alpha = 0.75,
        bins = [0, 50, 100, 150, 200, 250, 400])
plt.show()
```

```python
x = df_el['G_VRH']
y = df_el['K_VRH']
z = df_el['poisson_ratio']
plt.scatter(x, y, c = z, edgecolor= 'k', cmap = 'bwr')
plt.xlim(0, 300)
plt.show()
```

```python
import plotly.express as px
fig = px.scatter(df_el, x = 'G_VRH', y = 'K_VRH', color = 'poisson_ratio',
                color_continuous_scale = 'emrld', range_x = [0, 300])
fig.show()
```
