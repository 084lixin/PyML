```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image = mpimg.imread('2Dmini.png')
```

```python
image.shape
```

```python
image = image[:,:,:-1]
```

```python
plt.imshow(image)
plt.show()
```

```python
X = image.reshape(-1, 3)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
plt.imshow(kmeans.labels_.reshape(image[:,:,0].shape), cmap = 'brg')
plt.show()
```

```python
sum(kmeans.labels_ == 0) / len(kmeans.labels_)
```

```python
sum(kmeans.labels_ == 1) / len(kmeans.labels_)
```

```python
sum(kmeans.labels_ == 2) / len(kmeans.labels_)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clustering = AgglomerativeClustering(n_clusters = 3)
clustering.fit(X)
```

```python
clustering.labels_.shape
```

```python
image[:,:,0].shape
```

```python
plt.imshow(clustering.labels_.reshape(image[:,:,0].shape), cmap = 'brg')
plt.show()
```

```python
sum(clustering.labels_ == 0) / len(kmeans.labels_)
```

```python
sum(clustering.labels_ == 1) / len(kmeans.labels_)
```

```python
sum(clustering.labels_ == 2) / len(kmeans.labels_)
```
