```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 5)
y = 2 * x + 1
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, s = 200, c = 'w', marker = 's', edgecolor = 'k', linewidth = 3)
plt.show()
```

```python
plt.xlim(0, 30)

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.xlabel('x_label')
plt.ylabel('y_label')

plt.scatter(x, y, s = 200, c = 'r', marker = 'v')
plt.show()
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
plt.scatter(x, y, c='k')
plt.plot(x, y)
plt.show()
```
