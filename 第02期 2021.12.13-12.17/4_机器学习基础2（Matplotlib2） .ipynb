## 4.2 Matplotlib的使用2

#### 练习答案

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 50)
y = np.sin(x)
```

```python
plt.xlim(-5,5)
plt.plot(x, y)
```

#### 柱状图

```python
x = [1, 2, 3]
y = [5, 3, 4]
plt.bar(x, y)
```

```python
plt.bar(x, y, width = 0.2, color = 'r')
```

#### 绘制2D图

```python
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
X,Y = np.meshgrid(x,y)
Z = (1-X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)
plt.figure(figsize=(5,5))
plt.contourf(X,Y,Z)
plt.show()
```

```python
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
X,Y = np.meshgrid(x,y)
Z = (1-X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)
plt.figure(figsize=(5,5))
plt.pcolormesh(X,Y,Z)
plt.show()
```
