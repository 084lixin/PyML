## 5.4 决策树分类

```python
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
```

#### 导入事先生成的数据集

```python
f = open("knn.csv", encoding = 'UTF-8')
names = ['x', 'y', 'labels']
data = read_csv(f, names = names)
data.head()
```

#### 显示数据

```python
x1 = data['x'].iloc[:200,]
y1 = data['y'].iloc[:200,]
x2 = data['x'].iloc[200:400,]
y2 = data['y'].iloc[200:400,]
x3 = data['x'].iloc[400:600,]
y3 = data['y'].iloc[400:600,]
plt.figure(figsize=(8,8))
plt.scatter(x1,y1,c='b',marker='s',s=50)
plt.scatter(x2,y2,c='r', marker='^', s=50)
plt.scatter(x3,y3, c='g', s=50)
```

```python
excluded = ['labels']
X=data.drop(excluded, axis = 1)
y=data['labels'].values
# print(X)
# print(y)
```

#### 决策树分类

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

#### 结果可视化

```python
xx,yy = np.meshgrid(np.arange(0.2,1.31,0.01), np.arange(0.2,1.31,0.01))
```

```python
coords = np.c_[xx.ravel(), yy.ravel()]
```

```python
Z = clf.predict(coords)
print(xx.shape)
print(Z.shape)
Z = Z.reshape(xx.shape)
```

```python
#plt.contourf(xx, yy, Z, cmap = light_rgb)
plt.pcolormesh(xx, yy, Z)
plt.scatter(x1,y1,c='b',marker='s',s=50,alpha=0.8)
plt.scatter(x2,y2,c='r', marker='^', s=50, alpha=0.8)
plt.scatter(x3,y3, c='g', s=50, alpha=0.8)
plt.axis((0.2, 1.3,0.2,1.3))
```

```python
from matplotlib.colors import ListedColormap

light_rgb = ListedColormap([ '#AAAAFF', '#FFAAAA','#AAFFAA'])
plt.figure(figsize=(6,6))
plt.pcolormesh(xx, yy, Z, cmap = light_rgb)
plt.scatter(x1,y1,c='b',marker='s',s=50,alpha=0.8)
plt.scatter(x2,y2,c='r', marker='^', s=50, alpha=0.8)
plt.scatter(x3,y3, c='g', s=50, alpha=0.8)
plt.axis((0.2, 1.3,0.2,1.3))
```

## 5.5 KNN分类

```python
from sklearn import neighbors
clf2 = neighbors.KNeighborsClassifier(30)
clf2.fit(X, y)
```

```python
Z = clf2.predict(coords)
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(6,6))
plt.pcolormesh(xx, yy, Z, cmap = light_rgb)
plt.scatter(x1,y1,c='b',marker='s',s=50,alpha=0.8)
plt.scatter(x2,y2,c='r', marker='^', s=50, alpha=0.8)
plt.scatter(x3,y3, c='g', s=50, alpha=0.8)
plt.axis((0.2, 1.3,0.2,1.3))
```
