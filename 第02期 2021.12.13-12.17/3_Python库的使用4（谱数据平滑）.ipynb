## 3.4 实操：谱数据平滑

```python
import numpy as np
```

### 3.4.1 使用loadtxt读取

```python
with open('UV_noise.csv', mode = 'rb') as f:
    data = np.loadtxt(f,delimiter=",",skiprows=0)
data
```

```python
data.shape
```

```python
data = data.T
data
```

```python
wl = data[0]
Abs = data[1]
```

#### Savitzky-Golay平滑

```python
from scipy.signal import savgol_filter
Abs_smooth = savgol_filter(Abs, 9, 2)
```

### 3.4.2 绘图（课后阅读）

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,6))
plt.plot(wl, Abs)
plt.plot(wl, Abs_smooth)
plt.show()
```
